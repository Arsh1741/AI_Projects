```python
import pandas as pd
import re
```

```python
f = open('WhatsApp Chat with Watson hotel.txt','r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{1,2}\s[ap]m\s-'
```

```python
messages = re.split(pattern, data)[1:]
```

```python
messages
```

```python
dates = re.findall(pattern, data)
```

```python
dates
```

```python
#l=list(map(lambda x: x.replace(â€˜old_valueâ€™,â€™new_valueâ€™),l))
dates=list(map(lambda x: x.replace('\u202f',' '),dates))
```

```python
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
df['message_date'] = pd.to_datetime(df['message_date'], format="%d/%m/%y, %I:%M %p -")
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
df
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
users
```

```python
messages
```

```python
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```
